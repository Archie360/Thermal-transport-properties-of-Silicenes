```python
import scienceplots
import matplotlib.pyplot as plt
import numpy as np
plt.style.use(['science','notebook'])
import matplotlib.patches as mpatches
```

```python
e = [
        #  'hhk--4/hhk.bands.gnu',
        #  'hhk--2/hhk.bands.gnu',
         'hhk-0/hhk0.bands.gnu',
         'hhk-2/hhk-2.bands.gnu',
         'hhk-4/hhk.bands.gnu',
         'hhk-6/hhk.bands.gnu',
         ]
```

```python
p=1
plt.figure (figsize=(12,5))

for b in e:

    data = np.loadtxt(b)
    k = np.unique(data[:, 0])
    bands = np.reshape(data[:, 1], (-1, len(k)))

    gG1 = k[0]; M = k[30]; K = k[60]; gG2 = k[-1]

    efermi = max(bands[7 , :])

    plt.subplot(1,4,p)
    
    plt.axhline (0, c='gray', ls=':')
    plt.axvline(K, c='gray')
    plt.axvline(M, c='gray')
    for band in range(len(bands)):


        plt.plot(k, bands[band , :] -efermi, c='blue', lw=2)
        plt.xlabel('')
        plt.ylabel('E')
        plt.xlim(gG1 , gG2)
        plt.ylim (-2, 3 )
        plt.xticks ([gG1 , K, M, gG2], ['$\Gamma$ ', 'K', 'M','$\Gamma$ '])
        plt.tick_params(axis='x', which='minor', bottom=False , top=False)

    plt.axhline(y= efermi, c='gray', linestyle='--')
    # plt.axhline(y=bands[12 , :][100]-efermi, c='gray')
    # plt.axhspan(bands[12 , :][100]-efermi, bands[13 , :][100]-efermi, facecolor ='gray', alpha = .3)

    # plt.scatter(k[100], bands[13 , :][100], s=200, c='red' , alpha=.7)
    # plt.scatter(k[100], bands[12 , :][100], s=200, c='red' , alpha = .7)
    p+=1

plt.tight_layout()
plt.savefig('bands.png')
```
