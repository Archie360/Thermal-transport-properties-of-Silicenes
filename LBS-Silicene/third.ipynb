```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
```

```python
plt.style.use(['science','grid','notebook'])
```

```python
f0 = h5py.File("./lbs-0/third_order/kappa-m20201.hdf5")
f1 = h5py.File("./lbs-1/third_order/kappa-m20201.hdf5")
f2 = h5py.File("./lbs-2/third_order/kappa-m20201.hdf5")
f3 = h5py.File("./lbs-3/third_order/kappa-m20201.hdf5")
f4 = h5py.File("./lbs-4/third_order/kappa-m20201.hdf5")
```

```python
for a in list(f0):
    print(a,'.'*(25-len(a)),np.array(f0[a]).shape)
```

```python


plt.scatter(f0['temperature'][:],np.array(f0['kappa'])[:,0], lw = 2, c='b')
plt.scatter(f1['temperature'][:],np.array(f1['kappa'])[:,0], lw = 2, c='g')
plt.scatter(f2['temperature'][:],np.array(f2['kappa'])[:,0], lw = 2, c='y')
plt.scatter(f3['temperature'][:],np.array(f3['kappa'])[:,0], lw = 2, c='orange')
plt.scatter(f4['temperature'][:],np.array(f4['kappa'])[:,0], lw = 2, c='r')

    


plt.xlim(0,1000)
plt.ylim(0,32)
plt.ylabel('W/mK')
plt.xlabel('T')
```

```python
i=1
plt.figure(figsize=(12,2))
for f in [f0, f1,f2,f3,f4]:  

    kappa_xx_at_index_30 = f['mode_kappa'][30, :, :, 0].sum(axis=0)/ f['weight'][:].sum()
    plt.subplot(1,6,i)
    plt.bar([1,2,3,4,5,6],kappa_xx_at_index_30*100/sum(kappa_xx_at_index_30), 
                        edgecolor='black', lw=2,align='center', width=1, color = ['r','r','r','b', 'b', 'b'])
    i+=1
    plt.ylim(0,100)
    plt.xticks([1,2,3,4,5,6])
```
